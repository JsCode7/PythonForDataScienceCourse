```python
import pandas as pd
df = pd.read_csv("C:/Users/Jeste/Desktop/Data Science/base_datos_2008.csv/base_datos_2008.csv", nrows=1e6)
```

```python
#Con esta acción duplicaré el dataframe inicial
dfDuplicate = df.append(df)
```

```python
#Con esta acción reordenamos el dataframe sample(frac = 1)
dfDuplicate = dfDuplicate.sample(frac=1)
```

```python
#Guardamos el nuevo dataframe dropeando los duplicados en un df llamado dfclean
dfclean = dfDuplicate.drop_duplicates()
```

```python
#Como podemos ver el df clean es diferente del df inicial, esto debido a que en el inicial, 
#habian 4 datos que si estaban repetidos
len(dfclean) == len(df)
```

```python
#Con esto comprobamos la longitud del df
len(dfclean)
```

```python
#Con esta acción aplicaremos una eliminación de duplicados por los días del mes para hacer
#Un nuevo dataset, y de esta manera obtenemos un dataset solo con 31 dias del mes.
dfclean.drop_duplicates(subset = "DayofMonth")
```

```python
#Con esta función eliminamos del dataframe todos aquellas filas que tienen al menos un NaN
df.dropna()
```

```python
#pero le podemos aplicar un treshold para que mantengan filas con al menos 25 NaN
df.dropna(thresh = 25)
```

```python
#De la misma forma podemos hacer que mantengan un treshold de la longitud de columnas - 2 
df.dropna(thresh = len(df.columns) - 2)
```

```python
#Así como también podemos Hacer subset a partir de los CancellationCode que no tengan NaN
df.dropna(subset = ["CancellationCode"])
```

```python
#O también dejar solo los datos que no tienen un ArrTime NaN. 
df.dropna(subset = ["ArrTime"])
```
